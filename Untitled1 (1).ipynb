```python
import numpy as np
#Step 1 – Input signals weighted and combined as net input
#Step 2 – Net input fed into activation function
#Step 3 – Activation function output signal appropriately
#Step4  – Learning input signal weights based on prediction vs actuals
class Perceptron(object):
    
#  Learning rate needs to be defined   

    def __init__(self, n_iterations=50, random_state=1, learning_rate=0.05):
        self.n_iterations = n_iterations
        self.random_state = random_state
        self.learning_rate = learning_rate
        
#  updated of Weights is based on training examples.

  
    def fit(self,X, y):
        rgen = np.random.RandomState(self.random_state)
        self.coef_ = rgen.normal(loc=0.0, scale=0.01, size=1 + X.shape[1])
        for _ in range(self.n_iterations):
            for xi, expected_value in zip(X, y):
                predicted_value = self.predict(xi)
                self.coef_[1:] = self.coef_[1:] + self.learning_rate * (expected_value - predicted_value) * xi
                self.coef_[0] = self.coef_[0] + self.learning_rate * (expected_value - predicted_value) * 1
     
 #  calculate sun of all the weights 
    def net_input(self, X):
            dot=np.dot(X, self.coef_[1:])
            weightedSum= dot + self.coef_[0]
            return weightedSum
     
 # function for determine the outcome of the  weights 
    def activation_function(self, X):
            weightedSum = self.net_input(X)
            return np.where(weightedSum >= 0.0, 1, -1)
     
 # using activation_function to predict  
    def predict(self, X):
        return self.activation_function(X) 
     
 #calculate the score based on comparison of expected value & predicted value 
    def score(self, X, y):
        misclassified_data_count = 0
        for xi, target in zip(X, y):
            output = self.predict(xi)
            if(target != output):
                misclassified_data_count += 1
        total_data_count = len(X)
        self.score_ = (total_data_count - misclassified_data_count)/total_data_count
        return self.score_
```
